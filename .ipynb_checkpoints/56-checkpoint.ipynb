```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr

# Оригінальне зображення
im = cv2.imread('cat.jpg')
plt.axis('off')
plt.imshow(im)
plt.show()
```

```python
# Декомпозиція зображення по складових RGB-моделі - R складова
im_RGB = im.copy()
R = im_RGB.copy()

R[: , : , (1,2)] = 0

plt.axis('off')
plt.imshow(R)
plt.show()
```

```python
# Декомпозиція зображення по складових RGB-моделі - G складова
G = im_RGB.copy()

G[: , : , (0,2)] = 0

plt.axis('off')
plt.imshow(G)
plt.show()
```

```python
# Декомпозиція зображення по складових RGB-моделі - B складова
B = im_RGB.copy()

B[: , : , (0,1)] = 0

plt.axis('off')
plt.imshow(B)
plt.show()
```

```python
# Декомпозиція зображення по складових HSV-моделі - H складова
im_HSV = cv2.cvtColor(im_RGB, cv2.COLOR_RGB2HSV)
H = im_HSV.copy()

H[: , : , (1,2)] = 255

H_RGB = cv2.cvtColor(H, cv2.COLOR_HSV2RGB)
plt.axis('off')
plt.imshow(H_RGB)
plt.show()
```

```python
# Декомпозиція зображення по складових HSV-моделі - S складова
S = im_HSV.copy()

S[: , : , 0] = 179
S[: , : , 2] = 255

S_RGB = cv2.cvtColor(S, cv2.COLOR_HSV2RGB)
plt.axis('off')
plt.imshow(S_RGB)
plt.show()
```

```python
# Декомпозиція зображення по складових HSV-моделі - V складова
V = im_HSV.copy()

V[: , : , 0] = 179
V[: , : , 1] = 0

V_RGB = cv2.cvtColor(V, cv2.COLOR_HSV2RGB)
plt.axis('off')
plt.imshow(V_RGB)
plt.show()
```

```python
# Декомпозиція зображення по складових моделі CIE LAB  - L складова
im_LAB = cv2.cvtColor(im_RGB, cv2.COLOR_RGB2LAB)
L = im_LAB.copy()

L[: , : , (1, 2)] = 0

L_RGB = cv2.cvtColor(L, cv2.COLOR_LAB2RGB)
L_GRAY = cv2.cvtColor(L_RGB, cv2.COLOR_RGB2GRAY)
plt.axis('off')
plt.imshow(L_GRAY, cmap='gray')
plt.show()
```

```python
# Декомпозиція зображення по складових моделі CIE LAB  - A складова
A = im_LAB.copy()

A[: , : , 0] = 127
A[: , : , 2] = 0

A_RGB = cv2.cvtColor(A, cv2.COLOR_LAB2RGB)
A_GRAY = cv2.cvtColor(A_RGB, cv2.COLOR_RGB2GRAY)
cmap_a = clr.LinearSegmentedColormap.from_list('custom blue', ['Red', 'Grey', 'Green'], N=255)
plt.axis('off')
plt.imshow(A_GRAY, cmap=cmap_a)
plt.show()
```

```python
# Декомпозиція зображення по складових моделі CIE LAB  - B складова
B = im_LAB.copy()

B[: , : , 0] = 127
B[: , : , 1] = 0

B_RGB = cv2.cvtColor(B, cv2.COLOR_LAB2RGB)
B_GRAY = cv2.cvtColor(B_RGB, cv2.COLOR_RGB2GRAY)
cmap_b = clr.LinearSegmentedColormap.from_list('custom blue', ['Yellow', 'Grey', 'Blue'], N=255)
plt.axis('off')
plt.imshow(B_GRAY, cmap=cmap_b)
plt.show()
```
